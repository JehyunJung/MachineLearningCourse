# 알고리즘 결과 Visualization

## Import modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 손실 함수 정의

```python
def f(x):
    return 0.1 *x**4 -1.5*x**3 + 0.6*x**2 + 1.0*x +20.0
```

## 손실 함수 미분 정의

```python
def df(x):
    return 0.4 *x**3 -4.5*x**2 + 1.2*x + 1.0
```

```python
def d2f(x):
    return 1.2 *x**2 -9*x + 1.2
```

## Gradient Descent 알고리즘 구현

```python
def gradient_descent(initial_x, max_epoch=1000, learning_rate=0.01,eps=1e-5):
    x=initial_x
    x_log=[x]

    min_x=x
    min_y=f(min_x)

    for _ in range(max_epoch):
        grad=df(x)
        new_x=x-learning_rate*grad
        y=f(new_x)

        x_log.append(new_x)
        if min_y>y:
            min_x=new_x
            min_y=y

        if np.abs(x-new_x)<eps:
            break
        x=new_x
        
    return min_x,min_y,x_log
```

## 알고리즘 결과 시각화

```python
min_x1,min_y1,x_log1=gradient_descent(5)
min_x2,min_y2,x_log2=gradient_descent(-5)
min_x3,min_y3,x_log3=gradient_descent(0)
min_x4,min_y4,x_log4=gradient_descent(15)

y_log1=list(map(f,x_log1))
y_log2=list(map(f,x_log2))
y_log3=list(map(f,x_log3))
y_log4=list(map(f,x_log4))

plt.plot(x_log1,y_log1,'.-')
plt.plot(x_log2,y_log2,'.-')
plt.plot(x_log3,y_log3,'.-')
plt.plot(x_log4,y_log4,'.-')

```

```python
x=np.arange(-6,16,0.01)
dfy_log=list(map(df,x))
d2fy_log=list(map(d2f,x))
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12,6))

ax1.plot(x,dfy_log)
ax1.set(title="differential of loss function")

ax2.plot(x,d2fy_log)
ax2.set(title="secondary differential of loss function")

plt.show()

```
