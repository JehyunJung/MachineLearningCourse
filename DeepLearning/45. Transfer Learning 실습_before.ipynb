# Transfer Learning 실습

## Import modules

```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
```

## 하이퍼파라미터 정의

```python
EPOCHS=100
```

## 네트워크 구조 정의

```python
def MyModel():
    feat=tf.keras.applications.MobileNetV2(input_shape=(224,224,3),
                                          include_top=False) #뒷 부분의 fully connected layer은 제거한다
    feat.trainable=False #pre-trained model은 freeze
    
    seq=tf.keras.models.Sequential()
    seq.add(feat) # h*w*c
    seq.add(tf.keras.layers.GlobalAveragePooling2D()) #c
    seq.add(tf.keras.layers.Dense(1,activation='sigmoid'))
    return seq
```

## 데이터 불러오기 (Cats vs. Dogs)

```python
#splits = ("train[:80]", "train[81:]")
split=tfds.Split.TRAIN.subsplit(weighted=(8,2)) #train,test data 에서 split 진행해준다. (train,test을 8:2 비율로 분할하게 된다.)
dataset,meta=tfds.load('cats_vs_dogs',split=list(split),
                      with_info=True,as_supervised=True) #with_info -> 메타데이터 사용, as_supervised -> output data도 같이 가져온다.

train_ds,test_ds=dataset
```

```python
l2s=meta.features['label'].int2str
for img, label in test_ds.take(2):
    plt.figure()
    plt.imshow(img)
    plt.title(l2s(label))
    
train_ds
```

## 데이터 가공하기

```python
def preprocess(img,label):
    img=tf.cast(img,tf.float32)/255.0  #image 을 float32 로 casting 해서 normalize 진행
    img=tf.image.resize(img,(224,224)) #image 크기 조정   
    return img,label

train_ds=train_ds.map(preprocess).batch(32).prefetch(1024)
test_ds=test_ds.map(preprocess).batch(32).prefetch(1024)
```

## 모델 생성

```python
model=MyModel()
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

## 모델 학습

```python
model.fit(train_ds,validation_data=test_ds,epochs=10)
```
